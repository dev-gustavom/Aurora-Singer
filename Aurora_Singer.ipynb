# **Organização e descrição da telemetria**

**Temperatura interna e externa**

*Sensores medem a temperatura dentro do sistema e no ambiente externo. Para garantir que os componentes eletrónicos e mecânicos funcionam dentro de limites seguros.*

- temperatura interna: 18º - 27º
- temperatura externa: 40º - 50º

**Integridade estrutural (0/1)**

*Indicador binário que representa o estado da estrutura do veículo.*

1 → Estrutura íntegra

0 → Dano estrutural detectado

**Níveis de energia (%)**

*Indica a quantidade de energia disponível no sistema (baterias ou células de energia).*

- Minimo para lançamento: 85%

**Pressão dos tanques**

*Sensores monitorizam a pressão dos tanques de combustível ou gás. Para garantir pressão suficiente para funcionamento do sistema de propulsão.*

- Minimo: 90% da pressão nominal
- Maxímo: 110% da pressão nominal

**Status dos módulos críticos**

*Indica se subsistemas essenciais estão operacionais. Sistema de comunicação, Sistema de navegação, Sistema de propulsão*

- Para a decolagem todos devem estar OK.

# **PSEUDOCÓDIGO:**


Variáveis

temperatura_interna: inteiro

temperatura_externa: inteiro

integridade_estrutural: lógico

nivel_energia: inteiro

pressao_tanques: inteiro

status_modulos_criticos: lógico


Inicio

 Se temperatura_interna < 18 ou temperatura_interna > 27 então

    escreva "DECOLAGEM ABORTADA"

  Senão se temperatura_externa < 40 ou temperatura_externa > 50 então

    escreva "DECOLAGEM ABORTADA"

  Senão se integridade_estrutural == falso então

    escreva "DECOLAGEM ABORTADA"

  Senão se nivel_energia < 85 então

    escreva "DECOLAGEM ABORTADA"

  Senão se pressao_tanques < 90 ou pressao_tanques > 110 então

    escreva "DECOLAGEM ABORTADA"

  Senão se status_modulos_criticos == falso então

    escreva "DECOLAGEM ABORTADA"

 Senão

    escreva "PRONTO PARA DECOLAR"

Fim

```python
temperatura_interna = int(input("Temperatura interna: "))
temperatura_externa = int(input("Temperatura externa: "))
integridade_estrutural = bool(int(input("Integridade estrutural (1=OK,0=Falha): ")))
nivel_energia = int(input("Nível de energia (%): "))
pressao_tanques = int(input("Pressão dos tanques: "))
status_modulos_criticos = bool(int(input("Status módulos críticos (1=OK,0=Falha): ")))

if temperatura_interna < 18 or temperatura_interna > 27:
    print("DECOLAGEM ABORTADA")

elif temperatura_externa < 40 or temperatura_externa > 50:
    print("DECOLAGEM ABORTADA")

elif not integridade_estrutural:
    print("DECOLAGEM ABORTADA")

elif nivel_energia < 85:
    print("DECOLAGEM ABORTADA")

elif pressao_tanques < 90 or pressao_tanques > 110:
    print("DECOLAGEM ABORTADA")

elif not status_modulos_criticos:
    print("DECOLAGEM ABORTADA")

else:
    print("PRONTO PARA DECOLAR")
```
